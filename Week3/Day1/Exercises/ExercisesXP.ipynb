### Exercise 3 : Identifying Data Types


```python
df1.info()
```

**Quantitative**
* index, year, Avg hrs per day sleeping, standard error

**Qualitative**
* period, Type of data, age group, activity, sex

**Reason:** Quantitative columns are measurable numeric values, while qualitative columns describe categories or labels.
* dtype = object → Qualitative
* dtype = int64 or float64 → Quantitative

```python
df2.info()
```

**Quantitative**
* index, , anxiety disorder, drug use disorders, depression,Alcohol use disorders

**Qualitative**
* entity, code,year, schizophrenia, bipolar disorder, eating disorders

**Reason :**
* dtype = object → Qualitative
* dtype = int64 or float64 → Quantitative

```python
df3.info()
```

```python
#  Group columns
quantitative = []
qualitative = []

for col in df3.columns:
    if df3[col].dtype in ['int64','float64']:
        quantitative.append(col)
    else:
        qualitative.append(col)

print("Quantitative:", quantitative)
print("Qualitative:", qualitative)
```

**Reason :**
* dtype = object → Qualitative
* dtype = int64 or float64 → Quantitative

```python
print(df4.info())
```

**Quantitative**
* Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm

**Qualitative**
* Species

**Description of why :**
The Iris dataset has 150 rows of flower measurements. Quantitative columns (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) are numeric values describing dimensions, while the qualitative column (Species) is a categorical label. Numeric dtypes (float) are quantitative; object dtype is qualitative.

### Exercise 5: Basic Data Analysis with Google Colab

```python
#mean
df4["PetalLengthCm"].mean()
```

```python
#median
df4["PetalLengthCm"].median()
```

```python
#mode
df4["PetalLengthCm"].mode()[0]
```

```python
import matplotlib.pyplot as plt

df4.groupby('Species')['PetalLengthCm'].mean().plot(kind='bar', color='orange')
plt.title("Average Petal Length by Species")
plt.ylabel("Petal Length (cm)")
plt.show()

```

* The mean PetalLengthCm is about 3.75 cm, median ~4.35  cm.

* The histogram shows most PetalLength cluster between 2.5–5.8 cm.

* Average petal length differs clearly by species: Setosa has the smallest, Virginica the largest.
