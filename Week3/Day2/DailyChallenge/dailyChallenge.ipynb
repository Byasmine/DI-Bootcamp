```python
from PIL import Image,ImageOps
import matplotlib.pyplot as plt

# Load an image using PIL
image_path = 'flower-color-images/flower-color-images/flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
plt.axis("off")  # optional: hide axes
plt.show()
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
    return rotate(image, 30, axes=(1,0), reshape=True)
```

```python
# Show
plt.imshow(rotate_image_30_degrees(original_image))
plt.show()
```

```python
img_h = ImageOps.mirror(original_image)              # horizontal flip
img_v = ImageOps.flip(original_image)                # vertical flip

# Notebook:
display(img_h)
display(img_v)
```
